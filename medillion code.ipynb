```python
patients = spark.read.format("parquet").load(f"{base_path}/bronze/patients")
doctors = spark.read.format("parquet").load(f"{base_path}/bronze/doctors")
appointments = spark.read.format("parquet").load(f"{base_path}/bronze/appointments")
inventory = spark.read.format("parquet").load(f"{base_path}/bronze/inventory")
```

```python
# Cleanse bronze layer data and write to silver layer

from pyspark.sql.functions import col, trim, lower, when

# Cleanse patients: remove leading/trailing spaces, standardize gender, drop duplicates, filter nulls
patients_silver = (
    patients
    .withColumn("first_name", trim(col("first_name")))
    .withColumn("last_name", trim(col("last_name")))
    .withColumn("gender", lower(trim(col("gender"))))
    .filter(col("patient_id").isNotNull())
    .dropDuplicates(["patient_id"])
)

# Cleanse doctors: remove leading/trailing spaces, standardize specialization, drop duplicates, filter nulls
doctors_silver = (
    doctors
    .withColumn("doctor_name", trim(col("doctor_name")))
    .withColumn("specialization", lower(trim(col("specialization"))))
    .filter(col("doctor_id").isNotNull())
    .dropDuplicates(["doctor_id"])
)

# Cleanse appointments: filter out records with null patient_id or doctor_id, drop duplicates
appointments_silver = (
    appointments
    .filter(col("patient_id").isNotNull() & col("doctor_id").isNotNull())
    .dropDuplicates(["appointment_id"])
)

# Cleanse inventory: remove leading/trailing spaces, filter nulls, drop duplicates
inventory_silver = (
    inventory
    .withColumn("item_name", trim(col("item_name")))
    .filter(col("inventory_id").isNotNull())
    .dropDuplicates(["inventory_id"])
)

# Write cleansed data to silver layer as Parquet (or Delta if preferred)
patients_silver.write.mode("overwrite").parquet(f"{base_path}/silver/patients")
doctors_silver.write.mode("overwrite").parquet(f"{base_path}/silver/doctors")
appointments_silver.write.mode("overwrite").parquet(f"{base_path}/silver/appointments")
inventory_silver.write.mode("overwrite").parquet(f"{base_path}/silver/inventory")
```

```python
# Read cleansed silver layer data for further processing or analytics

patients_silver = spark.read.format("parquet").load(f"{base_path}/silver/patients")
doctors_silver = spark.read.format("parquet").load(f"{base_path}/silver/doctors")
appointments_silver = spark.read.format("parquet").load(f"{base_path}/silver/appointments")
inventory_silver = spark.read.format("parquet").load(f"{base_path}/silver/inventory")

display(patients_silver)
display(doctors_silver)
display(appointments_silver)
display(inventory_silver)
```

```python
#Join patients, appointments, and inventory to see what medicine each patient received in each appointment

from pyspark.sql.functions import col

# Note: appointments has 'prescription' (text) and inventory has 'item_name' (text)
# There is no inventory_id in appointments; joining on prescription == item_name as closest match

patient_medicine_appointments = (
    patients_silver
    .join(appointments_silver, on="patient_id", how="left")
    .join(inventory_silver, col("prescription") == col("item_name"), how="left")
    .select(
        col("patient_id"),
        col("first_name"),
        col("last_name"),
        col("appointment_id"),
        col("appointment_date"),
        col("item_name").alias("medicine_taken")
    )
)

display(patient_medicine_appointments)
```

```python
# Join patients, doctors, appointments, and inventory to display all information in one table

from pyspark.sql.functions import col

all_info = (
    patients_silver
    .join(appointments_silver, on="patient_id", how="left")
    .join(doctors_silver, on="doctor_id", how="left")
    .join(inventory_silver, col("prescription") == col("item_name"), how="left")
    .select(
        col("patient_id"),
        col("appointment_id"),
        col("doctor_id"),
        col("appointment_date"),
        col("consultation_fee"),
        col("status"),
        col("diagnosis"),
        col("prescription"),
        appointments_silver["last_updated"].alias("appointment_last_updated"),
        patients_silver["first_name"],
        patients_silver["last_name"],
        patients_silver["gender"],
        patients_silver["date_of_birth"],
        patients_silver["phone"].alias("patient_phone"),
        patients_silver["city"].alias("patient_city"),
        patients_silver["registration_date"],
        patients_silver["last_updated"].alias("patient_last_updated"),
        doctors_silver["doctor_name"],
        doctors_silver["specialization"],
        doctors_silver["experience_years"],
        doctors_silver["phone"].alias("doctor_phone"),
        doctors_silver["city"].alias("doctor_city"),
        doctors_silver["last_updated"].alias("doctor_last_updated")
    )
)

display(all_info)
```

```python
# Write the unified gold DataFrame (all_info) as a single file in a new folder under the silver layer

gold_path = f"{base_path}/silver/gold_analytics"

(
    all_info
    .coalesce(1)  # Ensure only one output file
    .write
    .mode("overwrite")
    .parquet(gold_path)
)
```

```python
gold_df = spark.read.format("parquet").load(gold_path)
display(gold_df)
```

```python
# Example next steps: Analyze gold data for insights

from pyspark.sql.functions import countDistinct, avg

# 1. Count unique patients, doctors, and appointments
summary_stats = gold_df.agg(
    countDistinct("patient_id").alias("unique_patients"),
    countDistinct("doctor_id").alias("unique_doctors"),
    countDistinct("appointment_id").alias("unique_appointments")
)
display(summary_stats)

# 2. Find top 5 most prescribed medicines
top_medicines = (
    gold_df.groupBy("prescription")
    .count()
    .orderBy(col("count").desc())
    .limit(5)
)
display(top_medicines)

# 3. Average number of appointments per doctor
appointments_per_doctor = (
    gold_df.groupBy("doctor_id", "doctor_name")
    .count()
    .agg(avg("count").alias("avg_appointments_per_doctor"))
)
display(appointments_per_doctor)
```
